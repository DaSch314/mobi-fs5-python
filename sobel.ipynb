```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 2.1.]()** Load and show the image:

```python
img_lena = np.array(plt.imread("data/lena.png"))
img_lena = img_lena.astype(float)
plt.imshow(img_lena, cmap='gray')
```

```python
from scipy.signal import convolve2d

def sobel_h(img):
    """Horizontal Sobel filter (gradient in x-direction)"""
    kernel = (1/8) * np.array([[1, 0, -1],
                               [2, 0, -2],
                               [1, 0, -1]], dtype=float)
    # Convert to grayscale if color image
    if img.ndim == 3:
        img_gray = img.mean(axis=2)
    else:
        img_gray = img
    return convolve2d(img_gray, kernel, mode="same", boundary="symm")

def sobel_v(img):
    """Vertical Sobel filter (gradient in y-direction)"""
    kernel = (1/8) * np.array([[1, 2, 1],
                               [0, 0, 0],
                               [-1, -2, -1]], dtype=float)
    # Convert to grayscale if color image
    if img.ndim == 3:
        img_gray = img.mean(axis=2)
    else:
        img_gray = img
    return convolve2d(img_gray, kernel, mode="same", boundary="symm")
```

**[Task 2.3.]()** Test your implementations from above:

```python
gx = sobel_h(img_lena)
gy = sobel_v(img_lena)


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(gx, cmap='gray')
plt.title("Partial derivative $g_x$ (horizontal edges)")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(gy, cmap='gray')
plt.title("Partial derivative $g_y$ (vertical edges)")
plt.colorbar()

plt.show()
```

**[Task 2.4.]()** Compute the image gradient magnitude:

```python
# Compute gradient magnitude
gradient_magnitude = np.sqrt(gx**2 + gy**2)

# Visualization
plt.figure(figsize=(10,8))
plt.imshow(gradient_magnitude, cmap='hot')
plt.title("Gradient Magnitude")
plt.colorbar(label='Magnitude')
plt.axis('off')
plt.tight_layout()
plt.show()
```
