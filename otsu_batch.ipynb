**[Task 2.1.]()** Implement Otsu thresholding as a re-usable function:

```python
def otsu_thresholding(imgf):
    img8 = (imgf * 255).round().astype(np.uint8)
    
    h = []
    for i in range (0, 256):
        counts = np.sum(img8 == i)
        h.append(counts)

    optimal_t = 0
    min_cost = np.inf

    for T in range (1, 256):
        class1 = img8[img8 <= T]
        class2 = img8[img8 > T]

        n1 = len(class1)
        n2 = len(class2)

        if n1 == 0 or n2 == 0:
            continue

        mu1 = np.mean(class1)
        mu2 = np.mean(class2)

        var1 = np.var(class1)
        var2 = np.var(class2)

        cost = n1 * var1 + n2 * var2
        if cost < min_cost:
            min_cost = cost
            optimal_t = T

            seg = (img8 > optimal_t).astype(np.uint8)
    
    return seg
```

**[Task 2.2.]()** Implement a re-usable function to compute the Dice coefficient:

```python
def compute_dice(bin1, bin2):
    if bin1.dtype != bool:
        bin1 = bin1 > 0
    if bin2.dtype != bool:
        bin2 = bin2 > 0
    
    schnitt = np.sum(bin1 & bin2)
    dice = 2 * schnitt / (np.sum(bin1) + np.sum(bin2))
    
    return dice
```

```python
img_O = plt.imread("data/NIH3T3/im/dna-0.png")
img_1 = plt.imread("data/NIH3T3/im/dna-1.png")

gt_0 = plt.imread("data/NIH3T3/gt/0.png")
gt_1 = plt.imread("data/NIH3T3/gt/1.png")

binary_0 = otsu_thresholding(img_O)
binary_1 = otsu_thresholding(img_1)

dice_0 = compute_dice(binary_0 > 0, gt_0 > 0)
dice_1 = compute_dice(binary_1 > 0, gt_1 > 0)

print("Dice Bild 0:", dice_0)
print("Dice Bild 1:", dice_1)
```

```python
indices = [28, 29, 33, 44, 46, 49]

dice_values = []

for i in indices: 
    img_i = plt.imread("data/NIH3T3/im/dna-" + str(i) + ".png")
    gt_i = plt.imread("data/NIH3T3/gt/" + str(i) + ".png")

    segmented_i = otsu_thresholding(img_i)
    dice_i = compute_dice(segmented_i, gt_i)

    dice_values.append(dice_i)
    print("Dice Bild", i, ":", dice_i)

average_dice = np.mean(dice_values)
print("Durchschnittlicher Dice-Koeffizient:", average_dice)
```
